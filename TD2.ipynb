#  6). Determine whether the following data support a proportionality argument for   y ∝ $Z^{\frac{1}{2}}$. If so, estimate the slope.

y ∝ $Z^{\frac{1}{2}}$

y=k $Z^{\frac{1}{2}}$

where k is the constant of proportionality

$y_{i+1}$=k $Z_{i+1}^{\frac{1}{2}}$

$y_i$=k $Z_i^{\frac{1}{2}}$

$\Delta y_{i+1} = y_{i+1}-y_{i} = k(Z_{i+1}^{\frac{1}{2}}-Z_{i}^{\frac{1}{2}})=k(\Delta Z_{i+1}^{\frac{1}{2}})$

$\Delta y_{1}=0$ , $\Delta Z_{1}^{\frac{1}{2}}=0$

Slope $k_{i+1}$ = $\frac {\Delta y_{i+1}}{\Delta Z_{i+1}^{\frac{1}{2}}}$

```python
df={"Y":[3.5,5,6,7,8],
    "Z":[ 3,6,9,12,15]}
df=pd.DataFrame(df)
a=[]
for i in range(len(df)):
    a.append(math.sqrt(df["Z"][i]))
df["sqrtZ"]=a
df
```

```python
a=[0]
b=[0]
for i in range(len(df)-1):
    a.append(df["sqrtZ"][i+1]-df["sqrtZ"][i])
    b.append(df["Y"][i+1]-df["Y"][i])
df['change_sqrtZ']=a
df['change_Y']=b
df
```

```python
df['slope']=round(df['change_Y']/df['change_sqrtZ'],2)
df
```

```python
fig = plt.figure(figsize =(10, 7))
plt.boxplot(df['slope'][1:])
plt.ylim(1,3)
plt.ylabel('Slope')
plt.title('Slope')
plt.show()
```

By Law of Large number.It approximate to normal distribution When Z to infinity. Average is more approximate to mean if there are not outlier.

```python
print('Mean = ',df['slope'][1:].mean())
print('Median = ',df['slope'][1:].median())
print('Mode = ',statistics.mode(df['slope'][1:]))
print('Standard Deviation = ',df['slope'][1:].std())
```

So Estimator of slope is 2.12.

```python
df['Model']=2.12*df['sqrtZ']
df
```

```python
plt.plot(df['sqrtZ'],df['Y'],label='Real')
plt.plot(df['sqrtZ'],df['Model'],label='Model')
plt.ylabel('Y')
plt.xlabel('sqrt(Z)')
plt.title('Real Vs Model')
plt.legend()
plt.show()
```

 # a). Force ∝ Stretch

F ∝ S

F=k S

where k is the constant of proportionality

$k={\frac{F}{S}}$

```python
df={"Force":[10, 20, 30, 40, 50, 60, 70, 80, 90],
    "Stretch":[ 19, 57, 94, 134, 173, 216, 256, 297, 343]}
df=pd.DataFrame(df)
df['K']=round(df["Force"]/df["Stretch"],2)
df
```

```python
fig = plt.figure(figsize =(10, 7))
plt.boxplot(df['K'])
plt.ylabel('K')
plt.title('K')
plt.show()
```

There is a outlier that is 0.53.

By Law of Large number.It approximate to normal distribution When number of data to infinity. And if data doesn't have outlier, Average is more approximate to mean. 

```python
print('Mean = ',round(df['K'][1:].mean(),2))
print('Median = ',round(df['K'][1:].median(),2))
print('Mode = ',statistics.mode(df['K'][1:]))
print('Standard Deviation = ',df['K'][1:].std())
```

So Estimator of K is 0.29 .

```python
df['Model']=0.29*df['Stretch']
df
```

```python
plt.plot(df['Stretch'],df['Force'],label='Real')
plt.plot(df['Stretch'],df['Model'],label='Model')
plt.ylabel('Force')
plt.xlabel('Stretch')
plt.title('Real Vs Model')
plt.legend()
plt.show()
```

# By the graph it seen good to fit porportionality model in this sample but it can have a big error for population.
